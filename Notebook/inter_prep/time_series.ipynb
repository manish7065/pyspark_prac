```python
import yfinance as yf
import datetime

# Define the date range
start = datetime.datetime(2020, 1, 1)
end = datetime.datetime.now()

# Fetch Tesla data from Yahoo Finance using yfinance
df_tesla = yf.download('TSLA', start=start, end=end)

# Display the data
df_tesla
```

## simple moving average

```python
df_tesla['Open'].plot(figsize = (16,9))
```

```python
from curses import window

from numpy import mean


df_tesla['Open: 10days rolling'] = df_tesla['Open'].rolling(window=10,min_periods=1).mean()
df_tesla
```

## Cummulative Moving Average


```python
## Expanding

df_tesla['Open'].expanding().mean().plot(figsize = (12,7))
```

## add more weight to recent data (EWMA- Exponential Weighted Moving Average)

```python
# smoothing factor = 0.1

df_tesla['EMA_0.1'] = df_tesla['Open'].ewm(alpha=0.1, adjust=False).mean()
```

```python
df_tesla['EMA_0.1'].plot(figsize=(12,7))
```

```python
df_tesla[["Open","EMA_0.1"]].plot(figsize=(15,8))
```

```python
df_tesla['EMA_0.3'] = df_tesla['Open'].ewm(span=5, adjust=False).mean()
df_tesla[["Open","EMA_0.3"]].plot(figsize=(15,8))

```

```python
df_tesla['EWMA_0.3'] = df_tesla['Open'].ewma(alpha=0.3, adjust=False).mean()
df_tesla[["Open","EWMA_0.3"]].plot(figsize=(15,8))

```
